```python
# Grouping and aggregating in pandas
# data pipeline, method-chaining
columns_list = ['job', 'purpose', 'credit_amount', 'age_years']
group_columns = ['job', 'purpose']

data_grouped_job_purpose = (
                                data[columns_list]
                                .groupby(group_columns)
                                .agg({"age_years": "max",
                                    "credit_amount": "mean"})
                                .reset_index()
                                .sort_values("credit_amount", ascending=False)
                                .rename(columns={"age_years": "max_age_year",
                                         "credit_amount": "mean_credit_amount"})
                            )
```

```python
# Assignment 5: Group the DataFrame based on the purpose column. The value should be the average credit amount per purpose.
selected_columns = ['purpose', 'credit_amount']
exclude_purposes = ['others', 'radio/television']

# Operators
'''
== : Gelijk aan
!= : Niet gelijk aan
<
>
'''

data_grouped = (data[selected_columns]
                .query("credit_amount >= 5000 & purpose != @exclude_purposes") # WHERE NOT IN
                .groupby('purpose')
                .mean()
                .reset_index()
                .sort_values(by="credit_amount", ascending=False)
)


data_grouped
```
