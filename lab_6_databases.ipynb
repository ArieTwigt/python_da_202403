```python
# import the required modules
import sqlite3
import pandas as pd
```

```python
# connect with the database
db_string = "data/adventureworks.db"

# create the connection
con = sqlite3.connect(db_string)
```

```python
# in python show the tables in the database
query = "SELECT name FROM sqlite_sequence"

result = con.execute(query)
```

```python
# specify a join query
query_sales_address = """
SELECT SalesOrderID, sales.OrderDate, address.City, address.CountryRegion
FROM SalesOrderHeader AS sales
JOIN Address AS address
  ON sales.ShipToAddressID = address.AddressID
"""

# run the query and add the result to a pandas DataFrame
df_sales_address = pd.read_sql_query(query_sales_address, con=con)

# show the first rows
df_sales_address.head()
```

```python
# query for getting the data from SalesOrderDetail
qry = """
SELECT SalesOrderID, OrderQty, LineTotal
FROM SalesOrderDetail
"""

df_sales_order_detail = pd.read_sql_query(qry, con=con)

df_sales_order_detail.head()
```

```python
# joining pandas DataFrames
df_sales_order_detail.merge(df_sales_address, 
                            how="left",
                            left_on="SalesOrderID",
                            right_on="SalesOrderID")
```
